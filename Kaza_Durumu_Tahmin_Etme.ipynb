```python
import numpy as np
from sklearn.naive_bayes import GaussianNB

X = np.array([
    [0, 0, 2, 0],  # Rain, bad, high, no
    [1, 1, 1, 1],  # snow, average, normal, yes
    [2, 0, 0, 0],  # clear, bad, light, no
    [2, 2, 0, 1],  # clear, good, light, yes
    [1, 2, 1, 0],  # snow, good, normal, no
    [0, 1, 0, 0],  # rain, average, light, no
    [0, 2, 1, 0],  # rain, good, normal, no
    [1, 0, 2, 0],  # snow, bad, high, no
    [2, 2, 2, 1],  # clear, good, high, yes
    [2, 0, 2, 1]   # clear, bad, high, yes
])

# Kazalar
Y = np.array([1, 1, 0, 0, 0, 0, 0, 1, 0, 1])
```

```python
model = GaussianNB()
model.fit(X, Y)
```

```python
# Yeni veri için tahmin:
prediction = model.predict([[2, 1, 0, 0]])  # clear, average, light, no
print("Tahmin edilen kaza durumu:", "Evet" if prediction[0] == 1 else "Hayır")
```
